# 第 1 步：请运行这个 cell 的函数

```python
# 基本模板
import datetime
from IPython.display import clear_output

# Edit contact information
def contactInfo(contact_people=['-'],contact_email=['-']):
    contac_info = ''
    for i in range(len(contact_people)):
        if contact_people[i]=='':
            contac_info = contac_info+contact_people[i]+contact_email[i]+', '
        else:
            contac_info = contac_info + 'Dr. '+contact_people[i]+' ('+contact_email[i]+'), '
        contac_info = contac_info[:-2]
    return contac_info
    pass

# Format UTC time
def getUtcDate():
    utc_time = datetime.datetime.utcnow()
    return utc_time.strftime('%Y-%m-%d %H:%M:%S')
    pass

# Edit database information
def dataBase(type_position = False,org_CN_name='-',org_EN_name='-',org_CN_country='-',org_EN_country='-',\
             direction='-',deadline='-',URL='-',\
             contact_people=['-'],contact_email=['-']):
    str = ''
    str = org_CN_name+'\n'+org_EN_name +'\n'+org_CN_country+'\n'+org_EN_country+'\n'
    contact = contactInfo(contact_people,contact_email)
    str = str + '<p>'+direction+r'; \nDeadline: '+deadline+r'; \nContact: '+contact+r'; \nURL: '+URL+'</p>\n\n'
    if type_position == 0:
        str = str + org_CN_country+org_CN_name+'招收博士研究生'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+' is recruiting for a Doctoral Student'+'\n'
    elif type_position == 1:
        str = str + org_CN_country+org_CN_name+'招收硕士研究生'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+" is recruiting for a Master's student"+'\n'
    elif type_position == 2:
        str = str + org_CN_country+org_CN_name+'招收博士研究生【和/或】硕士研究生'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+" is recruiting for a Doctoral student【and/or】a Master's student"+'\n'
    elif type_position == 3:
        str = str + org_CN_country+org_CN_name+'招聘博后'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+" is recruiting for a PostDoc"+'\n'
    else:
        str = str + '不是招聘信息'+'\n'
        
    utc_time = getUtcDate()
    str = str + 'UTC 时间：'+utc_time+'\n'
    return str
    pass

# Edit WeChat information
def weChat(type_position=False,org_CN_country='-',org_CN_name='-',direction='-',deadline='-',\
           contact_people=['-'],contact_email=['-'],URL='-'):
    str = ''
    if type_position == 0:
        str = org_CN_country+org_CN_name+' '+direction+' 方向 PhD 机会\n\n'
    elif type_position == 1:
        str = org_CN_country+org_CN_name+' '+direction+' 方向硕士机会\n\n'
    elif type_position == 2:
        str = org_CN_country+org_CN_name+' '+direction+' 方向硕士【和/或】博士机会\n\n'
    elif type_position == 3:
        str = org_CN_country+org_CN_name+' '+direction+' 方向 PostDoc 机会\n\n'
    else:
        str = '不是招聘信息'+'\n'
    str = str + '截止日期：'+deadline+'\n\n'
    contact_info = contactInfo(contact_people,contact_email)+'\n'
    str = str + '联系人：'+contact_info+'\n'
    str = str + '详情：'+URL+'\n\n'
    str = str + 'GISphere：【???】\n\n'
    if type_position == 0:
        str = str + '标签：'+org_CN_country+'；博士机会；【???】'
    elif type_position == 1:
        str = str + '标签：'+org_CN_country+'；硕士机会；【???】'
    elif type_position == 2:
        str = str + '标签：'+org_CN_country+'；博士机会；硕士机会；【???】'
    elif type_position == 3:
        str = str + '标签：'+org_CN_country+'；博后机会；【???】'
    else:
        str = str + '标签：'+org_CN_country+'；【???】'
    return str
    pass

# Edit WhatsApp information
def whatsApp(type_position=False,org_EN_country='-',org_EN_name='-',direction='-',deadline='-',\
           contact_people=['-'],contact_email=['-'],URL='-'):
    str = 'Direction: '+direction+'\n\n'
    if type_position == 0:
        str = str + 'PhD opportunity in '+org_EN_name+'\n\n'
    elif type_position == 1:
        str = str + "Master's opportunity in "+org_EN_name+'\n\n'
    elif type_position == 2:
        str = str + "PhD 【and/or】 Master's opportunity in "+org_EN_name+'\n\n'
    elif type_position == 3:
        str = str + 'PostDoc opportunity in '+org_EN_name+'\n\n'
    else:
        str = str + '不是招聘信息'+'\n'
    str = str + 'Deadline：'+deadline+'\n\n'
    contact_info = contactInfo(contact_people,contact_email)+'\n'
    str = str + 'Contact：'+contact_info+'\n'
    str = str + 'URL：'+URL+'\n\n'
    str = str + 'GISphere：【???】\n\n'
    if type_position == 0:
        str = str + 'Tags: '+org_EN_country+'; PhD opportunity;【???】'
    elif type_position == 1:
        str = str + 'Tags: '+org_EN_country+"; Master's opportunity;【???】"
    elif type_position == 2:
        str = str + 'Tags: '+org_EN_country+"; PhD 【and/or】 Master's opportunity;【???】"
    elif type_position == 3:
        str = str + 'Tags: '+org_EN_country+'; PostDoc opportunity;【???】'
    else:
        str = str + 'Tags: '+org_EN_country+';【???】'
    return str
    pass
```
