# 第 1 步：请运行这个 cell 的函数

```python
# 基本模板
import datetime
from IPython.display import clear_output

# Edit contact information
def contactInfo(contact_people=['-'],contact_email=['-']):
    contac_info = ''
    for i in range(len(contact_people)):
        if contact_people[i]=='':
            contac_info = contac_info+contact_people[i]+contact_email[i]+', '
        else:
            contac_info = contac_info + 'Dr. '+contact_people[i]+' ('+contact_email[i]+'), '
        contac_info = contac_info[:-2]
    return contac_info
    pass

def periodOfActivity(period_start=['-'],period_end=['-']):
    period = ''
    for i in range(len(period_end)):
        if period_start[i] == period_end[i] or period_end == '':
            period = period_start[i]
        else:
            period = period_start[i]+' - '+period_end[i]+', '
            period = period[:-2]
    return period
    pass

# Format UTC time
def getUtcDate():
    utc_time = datetime.datetime.utcnow()
    return utc_time.strftime('%Y-%m-%d %H:%M:%S')
    pass

# Edit database information
def dataBase(type_activity = False,org_CN_name='-',org_EN_name='-',org_CN_country='-',org_EN_country='-',\
             topic='-',deadline='-',URL='-',\
             contact_people=['-'],contact_email=['-'],period_start=['-'],period_end=['-']):
    str = ''
    str = org_CN_name+'\n'+org_EN_name +'\n'+org_CN_country+'\n'+org_EN_country+'\n'
    contact = contactInfo(contact_people,contact_email)
    str = str + '<p>'+topic+r'; \nDeadline: '+deadline+r'; \nContact: '+contact+r'; \nURL: '+URL+'<\p>\n'+'\n'
    if type_activity == 0:
        str = str + org_CN_country+org_CN_name+'举办暑期学校'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+' is hosting a Summer School'+'\n'
    elif type_activity == 1:
        str = str + org_CN_country+org_CN_name+'举办研讨会'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+" is hosting a Workshop"+'\n'
    elif type_activity == 2:
        str = str + org_CN_country+org_CN_name+'举办会议'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+" is is hosting a Conference"+'\n'
    elif type_activity == 3:
        str = str + org_CN_country+org_CN_name+'举办比赛'+'\n'
        str = str + org_EN_name+' in '+org_EN_country+" is hosting a Competition"+'\n'
    else:
        str = str + '不是学术活动'+'\n'
    period = periodOfActivity(period_start,period_end)
    str = str + period+'\n'
    utc_time = getUtcDate()
    str = str + 'UTC 时间：'+utc_time+'\n'
    return str
    pass

# Edit WeChat information
def weChat(type_activity=False,org_CN_country='-',org_CN_name='-',topic='-',deadline='-',\
           contact_people=['-'],contact_email=['-'],URL='-',period_start=['-'],period_end=['-']):
    str = ''
    if type_activity == 0:
        str = org_CN_country+org_CN_name+' 举办 '+topic+' 暑期学校\n\n'
    elif type_activity == 1:
        str = org_CN_country+org_CN_name+' 举办 '+topic+' 研讨会\n\n'
    elif type_activity == 2:
        str = org_CN_country+org_CN_name+' 举办 '+topic+' 学术会议\n\n'
    elif type_activity == 3:
        str = org_CN_country+org_CN_name+' 举办 '+topic+' 比赛\n\n'
    else:
        str = '不是活动信息'+'\n'
    str = str + '报名截止日期：'+deadline+'\n\n'
    period = periodOfActivity(period_start,period_end)
    str = str + '举办时间：'+period+'\n\n'
    contact_info = contactInfo(contact_people,contact_email)+'\n'
    str = str + '联系人：'+contact_info+'\n'
    str = str + '详情：'+URL+'\n\n'
    str = str + 'GISphere：【???】\n\n'
    if type_activity == 0:
        str = str + '标签：'+org_CN_country+'；暑期学校；【???】'
    elif type_activity == 1:
        str = str + '标签：'+org_CN_country+'；研讨会；【???】'
    elif type_activity == 2:
        str = str + '标签：'+org_CN_country+'；会议；【???】'
    elif type_activity == 3:
        str = str + '标签：'+org_CN_country+'；比赛；【???】'
    else:
        str = str + '标签：'+org_CN_country+'；【???】'
    return str
    pass

# Edit WhatsApp information
def whatsApp(type_activity=False,org_EN_country='-',org_EN_name='-',topic='-',deadline='-',\
           contact_people=['-'],contact_email=['-'],URL='-',period_start=['-'],period_end=['-']):
    str = 'Topic: '+topic+'\n\n'
    if type_activity == 0:
        str = str + org_EN_name+ 'Summer School opportunity\n\n'
    elif type_activity == 1:
        str = str +org_EN_name + 'Workshop opportunity\n\n'
    elif type_activity == 2:
        str = str + org_EN_name + 'Conference opportunity\n\n'
    elif type_activity == 3:
        str = str +org_EN_name+'Competition opportunity\n\n'
    else:
        str = str + '不是活动信息'+'\n'
    str = str + 'Deadline：'+deadline+'\n\n'
    period = periodOfActivity(period_start,period_end)
    str = str + 'Holding time：'+period+'\n\n'
    contact_info = contactInfo(contact_people,contact_email)+'\n'
    str = str + 'Contact：'+contact_info+'\n'
    str = str + 'URL：'+URL+'\n\n'
    str = str + 'GISphere：【???】\n\n'
    if type_activity == 0:
        str = str + 'Tags: '+org_EN_country+'; Summer School;【???】'
    elif type_activity == 1:
        str = str + 'Tags: '+org_EN_country+"; Workshop;【???】"
    elif type_activity == 2:
        str = str + 'Tags: '+org_EN_country+"; Conference;【???】"
    elif type_activity == 3:
        str = str + 'Tags: '+org_EN_country+'; Competition;【???】'
    else:
        str = str + 'Tags: '+org_EN_country+';【???】'
    return str
    pass
```
